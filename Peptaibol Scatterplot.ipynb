```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mat
```

```python
df = pd.read_csv('/mnt/c/Users/nawendt/Desktop/malaria/peptiabol.csv')
```

```python
df
```

```python
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(1,1)
yticks = [10,50,125,250]
x = df.loc[:,'Peptaibol Group'].values
y = df.loc[:,'Plasmodium EC50 (uM)'].values
c = df.loc[:,'Selectivity Index'].values
sct = ax.scatter(x, y, s=135,c=c, cmap="plasma", edgecolors='black',
            norm=mat.colors.PowerNorm(gamma=0.3))
ax.set_ylabel(r"$\it{Plasmodium}$ EC$_{50}$ ($\mu$M)")
ax.set_xlabel(r"Peptaibol Group")
cbax = plt.colorbar(sct)
cbax.ax.set_ylabel('Selectivity Index')
cbax.set_ticks(yticks)
plt.tight_layout()
```
